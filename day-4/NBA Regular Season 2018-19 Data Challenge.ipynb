# What was the average age of player in the league?

```python
import csv
import numpy as np
import pandas as pd


stats = pd.read_csv('../day-4/nbastats2018-2019.csv')
stats



meanAge = stats.describe().loc['mean']['Age']
print(f'The average age of a player in the league is {meanAge}')


# ANSWER -- 25.9


```

# What player scored the most points?

```python
stats

mostPoints = stats['Points']
mostPoints


names = stats['Name']
names

new_names = []

for name in range(len(names)):
    new_names.append(names[name])


zipped_values = list(zip(new_names, mostPoints))



sorted_points = sorted(zipped_values, key = lambda x: x[1], reverse = True)
print(sorted_points[0:1])



## Answer-- James Harden, 36.1
```

# Who played the most minutes during the season

```python
mostMinutes = stats.groupby(['Name', 'MP']).sum().sort_values('MP', ascending = False)

mostMinutes.head(3)

## ANSWER -- Bradley Beal, 36.9 MP
```

# Who is the highest paid player during the seasion?

```python
stats[['Name','Salary']].sort_values('Salary', ascending = False)

def makeInt(row):
    if row['Salary'] == '-':
        return 0
    else:
        return int(row['Salary'])
    

stats['int_sal'] = stats.apply(makeInt, axis = 1)
stats.sort_values('int_sal', ascending = False)

```
