```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
air_bnb = pd.read_csv('../weekend-project/AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighborhood groups are available and which shows up the most?
groups = air_bnb.groupby(['neighbourhood_group']).sum()

groups.sort_values('calculated_host_listings_count', ascending = False).head(1)


# ANSWER -- 5 groups, Manhattan shows up most with 277,073 host listings.
```

```python
# Are private rooms the most popular in manhattan?
air_bnb

popularRooms = air_bnb.groupby(['room_type', 'neighbourhood_group']).count().sort_values('neighbourhood', ascending = False)


popularRooms


## ANSWER -- No, Entire home/apartment are more popular.

```

```python
# Which hosts are the busiest and based on their reviews?

popularHost = air_bnb.groupby('host_name').sum()[['number_of_reviews']]

popularHost.sort_values('number_of_reviews', ascending = False)


# ANSWER -- Host Michael is busiest with 11,081 reviews.
```

```python
#Which neighorhood group has the highest average price?
averagePrice = air_bnb[['price','name', 'neighbourhood_group']]




neighbourhood = air_bnb['neighbourhood_group']



price = air_bnb['price']


newAve = []

for ave in range(len(price)):
    newAve.append(price[ave])


zipped_values = list(zip(newAve, neighbourhood))



sorted_price = sorted(zipped_values, key = lambda x: x[1], reverse = True)
print(sorted_price[0:1])



# ANSWER -- Staten Island with an average price of 70.
```

```python
# Which neighbor hood group has the highest total price?
air_bnb

totalPrice = air_bnb.groupby(['price','neighbourhood_group']).sum('price').sort_values('price', ascending = False).count


print(totalPrice)



# ANSWER -- Manhattan

```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby(['host_name', 'neighbourhood_group']).sum().sort_values('price', ascending = False)



#ANSWER --  Sonder(NYC), Blueground, Michael, David, Alex
```
